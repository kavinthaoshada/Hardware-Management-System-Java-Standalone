```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name).to(device)


def generate_response(prompt, model, tokenizer, max_length=200):
    inputs = tokenizer.encode(prompt, return_tensors='pt').to(device)
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response


def chat_with_llm():
    print("LLM: Hello! What is your name?")
    user_name = input("You: ")

    prompt = f"Hello! What is your name?\nUser: {user_name}\nLLM: "
    response = generate_response(prompt, model, tokenizer)
    print(f"LLM: Nice to meet you, {user_name}! How old are you?")

    user_age = input("You: ")
    prompt += f"Nice to meet you, {user_name}! How old are you?\nUser: {user_age}\nLLM: "
    response = generate_response(prompt, model, tokenizer)
    print(f"LLM: Thank you for sharing, {user_name}. Can you tell me your gender?")

    user_gender = input("You: ")
    prompt += f"Thank you for sharing, {user_name}. Can you tell me your gender?\nUser: {user_gender}\nLLM: "
    response = generate_response(prompt, model, tokenizer)
    print(f"LLM: How are you feeling today, {user_name}?")

    user_emotion = input("You: ")
    prompt += f"How are you feeling today, {user_name}?\nUser: {user_emotion}\nLLM: "
    response = generate_response(prompt, model, tokenizer)
    print(f"LLM: Do you have any mental discomfort or concerns you want to talk about?")

    user_discomfort = input("You: ")
    prompt += f"Do you have any mental discomfort or concerns you want to talk about?\nUser: {user_discomfort}\nLLM: "
    response = generate_response(prompt, model, tokenizer)

    positive_feedback = "Thank you for sharing your thoughts and feelings. Remember, it's always good to talk about how you feel. You are strong and capable!"
    print(f"LLM: {positive_feedback}")


if __name__ == "__main__":
    chat_with_llm()
```
